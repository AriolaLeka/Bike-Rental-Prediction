#One Hot Encoding

```python
 # one-hot-encoding for the categorical columns 
def hot_encoding(dataset_forhe, testing=False):
        train_data = dataset_forhe.copy()

        if "weathersit" in train_data_m1 and 'mnth' in train_data_m1 and 'season'in train_data_m1:

              weathersit_dummies = pd.get_dummies(train_data['weathersit'], prefix='weathersit')
              season_dummies = pd.get_dummies(train_data['season'], prefix='season')
              #year_dummies = pd.get_dummies(train_data['yr'], prefix='yr')
              month_dummies = pd.get_dummies(train_data['mnth'], prefix='mnth')
              hour_dummies = pd.get_dummies(train_data['hr'], prefix='hr')
              weekday_dummies = pd.get_dummies(train_data['weekday'], prefix='weekday')
              #holiday_dummies = pd.get_dummies(train_data['holiday'], prefix='holiday')
              #workingday_dummies = pd.get_dummies(train_data['workingday'], prefix='workingday')


              train_data = pd.concat([train_data, season_dummies], axis=1)
              train_data = pd.concat([train_data, weathersit_dummies], axis=1)
              #train_data = pd.concat([train_data, year_dummies], axis=1)
              train_data = pd.concat([train_data, month_dummies], axis=1)
              train_data = pd.concat([train_data, hour_dummies], axis=1)
              train_data = pd.concat([train_data, weekday_dummies], axis=1)
              #train_data = pd.concat([train_data, holiday_dummies], axis=1)
              #train_data = pd.concat([train_data, workingday_dummies], axis=1)



              train_data.drop('weathersit', axis=1, inplace=True)
              train_data.drop('season', axis=1, inplace=True)
              #train_data.drop('yr', axis=1, inplace=True)
              train_data.drop('mnth', axis=1, inplace=True)
              train_data.drop('hr', axis=1, inplace=True)
              train_data.drop('weekday', axis=1, inplace=True)
              #train_data.drop('holiday', axis=1, inplace=True)
              #train_data.drop('workingday', axis=1, inplace=True)

              if testing == True:
                train_data['mnth_1'] = 0
                train_data['mnth_2'] = 0
                train_data['mnth_3'] = 0
                train_data['mnth_4'] = 0
                train_data['mnth_5'] = 0
                train_data['mnth_6'] = 0
                train_data['mnth_7'] = 0
                train_data['weathersit_4'] = 0
                train_data['season_2'] = 0

        else:
          hour_dummies = pd.get_dummies(train_data['hr'], prefix='hr')
          weekday_dummies = pd.get_dummies(train_data['weekday'], prefix='weekday')

          train_data = pd.concat([train_data, hour_dummies], axis=1)
          train_data = pd.concat([train_data, weekday_dummies], axis=1)

          train_data.drop('hr', axis=1, inplace=True)
          train_data.drop('weekday', axis=1, inplace=True)



        return train_data

```
